# ETH/MATIC CALCULATION

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import statistics

pair1 = 'ETH-USD'
pair2 = 'MATIC-USD'

data_pair1 = yf.download(pair1, start='2022-11-01', end='2023-02-14')
data_pair2 = yf.download(pair2, start='2022-11-01', end='2023-02-14')

pair1_close = data_pair1['Close']
pair2_close = data_pair2['Close']

ratio = pair1_close / pair2_close

df_ratio = pd.DataFrame({'Ratio': ratio})

plt.figure(figsize=(10, 6))
plt.plot(df_ratio.index, df_ratio['Ratio'])
plt.title(f'{pair1}/{pair2} Ratio')
plt.xlabel('Date')
plt.ylabel('Ratio')
plt.grid(True)
plt.show()
```

# Mean & Standard Deviation Calculation

```python
# Mean Cal
mean_ratio = statistics.mean(df_ratio['Ratio'])
# Std Dev Cal
std_dev_ratio = statistics.stdev(df_ratio['Ratio'])
#Std Dev -1
std_dev_min1 = mean_ratio-std_dev_ratio
#std Dev -2
std_dev_min2 = std_dev_min1 - std_dev_ratio
#Std Dev +1
std_dev_plus1 = mean_ratio+std_dev_ratio
#Std Dev +2
std_dev_plus2 = std_dev_plus1 + std_dev_ratio
```

# Summary

```python
print("Mean:                  ",mean_ratio)
print("Standard Deviation:    ",std_dev_ratio)
print("Standard Deviation -2: ",std_dev_min2)
print("Standard Deviation -1: ",std_dev_min1)
print("Standard Deviation +1: ",std_dev_plus1)
print("Standard Deviation +2: ",std_dev_plus2)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_ratio.index, df_ratio['Ratio'])
plt.title(f'{pair1}/{pair2} Ratio')
plt.xlabel('Date')
plt.ylabel('Ratio')
plt.grid(True)
plt.axhline(y=mean_ratio, color='red', linestyle='--')
plt.axhline(y=std_dev_min1, color='green', linestyle='--')
plt.axhline(y=std_dev_min2, color='yellow', linestyle='--')
plt.axhline(y=std_dev_plus1, color='blue', linestyle='--')
plt.axhline(y=std_dev_plus2, color='purple', linestyle='--')

plt.show()
```
